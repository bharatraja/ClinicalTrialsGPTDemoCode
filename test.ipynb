```python
def camel_case_split(str):
    #first word will 
    words = [[str[0].upper()]]

 
    for c in str[1:]:
        if words[-1][-1].islower() and c.isupper():
            words.append(list(c))
        else:
            words[-1].append(c)
 
    words= [''.join(word) for word in words]
    return ' '.join(words)
    #print(' '.join(words))  

# Driver code
str = "goodGeeks"
print(camel_case_split(str))
```
